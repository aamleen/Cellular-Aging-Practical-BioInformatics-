```python
uploaded=files.upload()
cell_df3=pd.read_csv(io.BytesIO(uploaded['dataset3.csv']),encoding="ISO-8859-1")
cell_df3
```

```python
cell_df1.drop(['Source'],axis=1,inplace=True,errors='ignore')
cell_df2.drop(['Sources'],axis=1,inplace=True,errors='ignore')
cell_df3.drop(['Sources'],axis=1,inplace=True,errors='ignore')
cell_df1.dropna(how='all',axis=0,inplace=True)
cell_df2.dropna(how='all',axis=0,inplace=True)
cell_df3.dropna(how='all',axis=0,inplace=True)
```

```python

display(cell_df1)
display(cell_df2)
display(cell_df3)
```

```python
cell_df3.rename(columns={'Replicative Lifespan (d)':'Average Cell Lifespan (d)'},inplace=True,errors="ignore")
combined_celllife_df=pd.concat([cell_df2[['Group','Species','Cell Type','Average Cell Lifespan (d)','Real Life Span']],
                              cell_df3[['Group','Species','Cell Type','Average Cell Lifespan (d)','Real Life Span']]
                              ], ignore_index=True)
combined_celllife_df
```

```python
sns.set_style('darkgrid')
figure,axes=plt.subplots(1,2,figsize=(24,12))
sns.boxplot(ax=axes[0],x='Group',y='Average Cell Lifespan (d)',data=cell_df1)
sns.boxplot(ax=axes[1],x='Group',y='Real Life Span',data=cell_df1)
```

```python
figure,axes=plt.subplots(3,1,figsize=(24,12))
total_combined_celllife_df=pd.concat([cell_df1,combined_celllife_df.drop(['Cell Type'],axis=1)],ignore_index=True)
sns.scatterplot(ax=axes[0],x='Average Cell Lifespan (d)',y='Real Life Span',data=cell_df1,hue='Group',s=100).set(xlim=(0,400),ylim=(0,50))
sns.scatterplot(ax=axes[1],x='Average Cell Lifespan (d)',y='Real Life Span',
                data=total_combined_celllife_df, hue='Group',s=100).set(xlim=(0,400),ylim=(0,50))
sns.scatterplot(ax=axes[2],x='Average Cell Lifespan (d)',y='Real Life Span',data=combined_celllife_df,hue='Cell Type',s=100).set(xlim=(0,220),ylim=(0,50))
```

```python
plt.figure(figsize=(24,12))
total_combined_celllife_df[['Average Cell Lifespan (d)','Real Life Span']].corr(method="pearson")
i=0
j=0
sns.relplot(x="Average Cell Lifespan (d)",y='Real Life Span',col='Group',data=total_combined_celllife_df,kind='line')
"""
for group in total_combined_celllife_df['Group'].unique():
  temp_df=total_combined_celllife_df.loc[total_combined_celllife_df['Group']==group]
  sns.plot(ax=axes[(i++)%2][j],x='Average Cell Lifespan (d)',y='Real Life SPan',data=''
  if(i>1):
    j=0"""
```

```python
mammals_df=total_combined_celllife_df.loc[total_combined_celllife_df['Group']=='mammals']
display(mammals_df)
display(mammals_df.corr(method="pearson"))
fig,axes=plt.subplots(1,2,figsize=(24,12))
sns.boxplot(ax=axes[0],x='Group',y='Real Life Span',data=mammals_df)
sns.boxplot(ax=axes[1],x='Group',y='Average Cell Lifespan (d)',data=mammals_df)
```

```python
fig,axes=plt.subplots(3,1,figsize=(24,12))
axes[0].set_xlabel('Average Cell Lifespan (d)')
axes[1].set_xlabel('Average Cell Lifespan (d)')
axes[0].set_ylabel('Real Life Span')
axes[1].set_ylabel('Real Life Span')
axes[0].plot(mammals_df['Average Cell Lifespan (d)'],mammals_df['Real Life Span'],'o--r')
axes[1].plot(mammals_df['Average Cell Lifespan (d)'],mammals_df['Real Life Span'],'o--r')
plt.xlim([0,500])
```
